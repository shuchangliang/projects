## The Boston Housing Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
boston = pd.read_csv('datasets_35807_51165_Boston.csv')
boston.head()
```

```python
boston.drop(columns = 'Unnamed: 0', inplace=True)
```

```python
boston.head()
```

```python
boston.shape
```

```python
boston.describe()
```

```python
boston.isnull().sum() # No missing values
```

```python
# Relationship between # of Rooms and Price
plt.figure(figsize=(14, 8))
sns.regplot(x='rm', y='medv', data=boston, fit_reg=True)
plt.title('Relationship between # of Rooms and Price', fontsize=12)
plt.xlabel('# of Rooms',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.show()
```

It is very obvious that the # of room increases, the price also increases.

```python
# Relationship between % Lower Status of Population and Price
plt.figure(figsize=(14, 8))
sns.regplot(x='lstat', y='medv', data=boston, fit_reg=True)
plt.title('Relationship between % Lower Status of Population and Price', fontsize=12)
plt.xlabel('% Lower Status of Population',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.show()
```

It is reasonable that the % Lower Status of Population has negtive impact on house prices.

```python
# Relationship between Concentration and Price
plt.figure(figsize=(14, 8))
sns.regplot(x='nox', y='medv', data=boston, fit_reg=True)
plt.title('Relationship between Nitric Oxides Concentration and Price', fontsize=12)
plt.xlabel('Nitric Oxides Concentration',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.show()
```

Nitric Oxides is bad for human body, so it has a negative impact on house prices.

```python
# Relationship between Weighted Distances and Price
plt.figure(figsize=(14, 8))
sns.regplot(x='dis', y='medv', data=boston, fit_reg=True)
plt.title('Relationship between Weighted Distances and Price', fontsize=12)
plt.xlabel('Weighted Distance',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.show()
```

```python
# Relationship between Crime Rate and Price
plt.figure(figsize=(14, 8))
sns.regplot(x='crim', y='medv', data=boston, fit_reg=True)
plt.title('Relationship between Crime Rate and Price', fontsize=12)
plt.xlabel('Crime Rate',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.show()
```

People tend to live in an area with low crime rate, so the lower the cirme rate, the higher the house price.

```python
boston_selected_var_df = boston[['crim','nox','rm','dis','ptratio','lstat','medv']]
boston_selected_var_df
```

```python
# find the correlation between each variable
boston_selected_var_df.corr()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Ordinary least squares model
model = ols('medv ~ crim + nox + rm + dis + ptratio + lstat', boston_selected_var_df).fit()
```

```python
print(model.summary())
```

```python
prediceted_prices = model.fittedvalues
prediceted_prices
```

```python
# Original Price vs. Ordinary least squares model Predicted Price
plt.figure(figsize=(14, 8))
sns.regplot(x=boston_selected_var_df.medv, y=prediceted_prices, data=boston, fit_reg=True)
plt.title('Original Price vs. Predicted Price', fontsize=12)
plt.xlabel('Original Price',fontsize=12)
plt.ylabel('Predicted Price',fontsize=12)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_error_for_ols_model = np.sqrt(mean_squared_error(boston_selected_var_df.medv, prediceted_prices))
mean_error_for_ols_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = boston_selected_var_df.drop('medv', axis=1)
y = boston_selected_var_df['medv']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinReg = LinearRegression()
```

```python
# linear regression model
LinReg.fit(x_train, y_train)
```

```python
y_pred = LinReg.predict(x_test)
```

```python
y_pred
```

```python
# Original Price vs. linear regression model Predicted Price
plt.figure(figsize=(14, 8))
sns.regplot(x=y_test, y=y_pred, data=boston, fit_reg=True)
plt.title('Original Price vs. Predicted Price', fontsize=12)
plt.xlabel('Original Price',fontsize=12)
plt.ylabel('Predicted Price',fontsize=12)
plt.show()
```

```python
mean_error_for_LinReg_model = np.sqrt(mean_squared_error(y_test, y_pred))
mean_error_for_LinReg_model
```
